```python
from dateutil.parser import parse
import pandas as pd
import numpy as np
import ast
```

```python
AA_Meets = pd.read_csv('Meeting_Schedule.csv')
AA_Meets.head(3)
```

```python
AA_Meets['Schedule'] = [ast.literal_eval(AA_Meets['Timings'][i]) for i in range(len(AA_Meets['Timings']))]
AA_Meets = AA_Meets.drop(['Timings'], axis=1)
```

```python
AA_Meets['Year'] = [parse(AA_Meets['Published_Date'][i], fuzzy=True).year for i in range(len(AA_Meets['Published_Date']))]
AA_Meets['Meetings_Held'] = [len(item) for item in AA_Meets['Schedule']]
AA_Meets.head()
```

```python
location = []
for i in range(len(AA_Meets['Headline'])):
    if 'engagements' in AA_Meets['Headline'][i].lower():
        location.append('engagements')
    elif 'chennai' in AA_Meets['Headline'][i].lower():
        location.append('chennai')
    elif 'madurai' in AA_Meets['Headline'][i].lower():
        location.append('madurai')
    elif 'thiruvananthapuram' in AA_Meets['Headline'][i].lower():
        location.append('thiruvananthapuram')
    elif 'tirunelveli' in AA_Meets['Headline'][i].lower():
        location.append('tirunelveli')
    elif 'coimbatore' or 'caoimbatore' in AA_Meets['Headline'][i].lower():
        location.append('coimbatore')
    else:
        location.append(None)
```

```python
# Checking the None coount in list
print(sum(x is None for x in location))
```

```python
AA_Meets['Location'] = location
AA_Meets = AA_Meets.drop(['Headline'], axis=1)
AA_Meets.head()
```

```python
AA_Meets.to_csv('AA_Meets.csv', index=False)
```
