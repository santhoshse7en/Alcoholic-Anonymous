```python
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_binary
import pandas as pd
import time
```

```python
class aa_meetups:
    
    def __init__(self, region):
        self.region = region
        self.url = 'http://www.aagsoindia.org/'
        options = Options()
        options.headless = True
        options=options
        driver = webdriver.Chrome(options=options)
        driver.get(self.url)

        # Select area from drop-down menu
        select = driver.find_element_by_id('address').send_keys(str(self.region))
        button = driver.find_element_by_css_selector('li.second-search-input.front_search').click()
        time.sleep(3)
        
        soup = BeautifulSoup(driver.page_source, 'lxml')
        driver.quit()
        
        self.aa_meetings_dataframe = pd.DataFrame(columns = ['Group Name', 'Address', 'Town / City', 'State', 'Zipcode', 'Phone', 'Time'])

        try:
            meetings = soup.select('.list-content')

            for meets in meetings:
                try:
                    self.aa_meetings_dataframe.loc[len(self.aa_meetings_dataframe)] = [meets.select_one('.loc-name').text.strip(),
                                                                                       meets.select_one('.loc-addr').text.strip(),
                                                                                       meets.select_one('.loc-addr3').text.split(',')[0].strip(),
                                                                                       meets.select_one('.loc-addr3').text.split(',')[-1].strip()[:-8],
                                                                                       meets.select_one('.loc-addr3').text.split(',')[1].strip()[-6:],
                                                                                       meets.select_one('.loc-phone').text.strip(),
                                                                                       meets.select_one('.loc-hours1').text.strip()]
                except:
                    pass
        except:
            pass
        
    def create_csv(self):
        self.aa_meetings_dataframe.to_csv(str(self.region)+'.csv', index=False)
        return print(str(self.region)+'.csv ' + 'successfully created')
```

```python
meetups = aa_meetups('EAST DELHI')
```
