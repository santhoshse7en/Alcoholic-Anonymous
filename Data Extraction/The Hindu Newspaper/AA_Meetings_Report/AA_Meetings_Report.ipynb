```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os, sys
%matplotlib notebook
```

```python
AA_Meets = pd.read_csv('AA_Meets.csv')
AA_Meets.head()
```

```python
y_2019 = AA_Meets[AA_Meets['Year']==2019]
y_2018 = AA_Meets[AA_Meets['Year']==2018]
y_2017 = AA_Meets[AA_Meets['Year']==2017]
y_2016 = AA_Meets[AA_Meets['Year']==2016]
y_2015 = AA_Meets[AA_Meets['Year']==2015]
y_2014 = AA_Meets[AA_Meets['Year']==2014]
y_2013 = AA_Meets[AA_Meets['Year']==2013]
y_2012 = AA_Meets[AA_Meets['Year']==2012]
y_2011 = AA_Meets[AA_Meets['Year']==2011]
y_2010 = AA_Meets[AA_Meets['Year']==2010]
```

```python
# AA Meetings held in weekdays during time-interval (2010-2019)
df = pd.DataFrame({'2019': pd.value_counts(y_2019['Day']),
                   '2018': pd.value_counts(y_2018['Day']),
                   '2017': pd.value_counts(y_2017['Day']),
                   '2016': pd.value_counts(y_2016['Day']),
                   '2015': pd.value_counts(y_2015['Day']),
                   '2014': pd.value_counts(y_2014['Day']),
                   '2013': pd.value_counts(y_2013['Day']), 
                   '2012': pd.value_counts(y_2012['Day']),
                   '2011': pd.value_counts(y_2011['Day']),
                   '2010': pd.value_counts(y_2010['Day']),})

df = df.fillna(0).astype(int)
```

```python
%%time
# AA Meetings in Chennai during time interval (2010-2019)
weedays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
for index, item in enumerate(weedays):
    x = df.loc[str(item)].index.tolist()
    y = df.loc[str(item)].tolist()

    fig, ax = plt.subplots()    
    width = 0.4 # the width of the bars 
    ind = np.arange(len(y))  # the x locations for the groups
    ax.barh(ind, y, width)
    ax.set_yticks(ind+width/2)
    ax.set_yticklabels(x, minor=False)
    for i, v in enumerate(y):
        ax.text(v-0.5, i +0.1, " "+str(v), size = 10)
    plt.title('AA Meetings held chennai with time interval 2010 to 2019')
    plt.xlabel('Total number of Meetings held on ' +str(item))
    plt.ylabel('Time Interval')  
    plt.savefig(os.path.join("D:\\!Code!\\CDRI\\The Hindu\\AA_Meetings_Report\\Weekdays_Visuals\\"+str(item)+".png"), dpi=1000, format="png", bbox_inches="tight") # use format='svg' or 'pdf' for vectorial pictures
    
    sys.stdout.write('\r' + str(index) + ' : ' + str(item) + '\r')
    sys.stdout.flush()
```

```python
# AA Meetings held in weekdays during time-interval (2010-2019)
df = pd.DataFrame({'2019': pd.value_counts(y_2019['Location']),
                   '2018': pd.value_counts(y_2018['Location']),
                   '2017': pd.value_counts(y_2017['Location']),
                   '2016': pd.value_counts(y_2016['Location']),
                   '2015': pd.value_counts(y_2015['Location']),
                   '2014': pd.value_counts(y_2014['Location']),
                   '2013': pd.value_counts(y_2013['Location']), 
                   '2012': pd.value_counts(y_2012['Location']),
                   '2011': pd.value_counts(y_2011['Location']),
                   '2010': pd.value_counts(y_2010['Location']),})
df = df.fillna(0).astype(int)
```

```python
%%time
# AA Meetings in Chennai during time interval (2010-2019)
locations = ['chennai', 'coimbatore', 'engagements', 'madurai']
for index, item in enumerate(locations):
    x = df.loc[str(item)].index.tolist()
    y = df.loc[str(item)].tolist()

    fig, ax = plt.subplots()    
    width = 0.4 # the width of the bars 
    ind = np.arange(len(y))  # the x locations for the groups
    ax.barh(ind, y, width)
    ax.set_yticks(ind+width/2)
    ax.set_yticklabels(x, minor=False)
    for i, v in enumerate(y):
        ax.text(v-0.5, i +0.1, " "+str(int(v)), size = 10)
    plt.title('AA Meetings held chennai with time interval 2010 to 2019')
    plt.xlabel('Total number of Meetings held in '+str(item))
    plt.ylabel('Time Interval')      
    plt.savefig(os.path.join("D:\\!Code!\\CDRI\\The Hindu\\AA_Meetings_Report\\Location_Visuals\\"+str(item)+".png"), dpi=1000, format='png', bbox_inches='tight') # use format='svg' or 'pdf' for vectorial pictures
    
    sys.stdout.write('\r' + str(index) + ' : ' + str(item) + '\r')
    sys.stdout.flush()
```
